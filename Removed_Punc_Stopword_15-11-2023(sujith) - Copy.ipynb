```python
sen=sent_tokenize(contents)
sen
```

```python
wor=word_tokenize(contents)
wor
```

```python
import string
for punc in string.punctuation:
    contents=contents.replace(punc,'')
print(contents)
```

```python
from nltk.corpus import stopwords
```

```python
stop_words=set(stopwords.words('english'))
cleaned=[]
for word in wor:
    if word.lower() not in stop_words:
        cleaned.append(word)
print(cleaned)
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem.snowball import SnowballStemmer

stemmed=[PorterStemmer().stem(w) for w in cleaned]
print("Stemming in PorterStemmer")
print(stemmed)

stemmed=[LancasterStemmer().stem(w) for w in cleaned]
print("Stemming in LancasterStemmer")
print(stemmed)

stemmed=[SnowballStemmer('english').stem(w) for w in cleaned]
print("Stemming in SnowballStemmer")
print(stemmed)
```
