```python
new_iris=iris[['petal_width','petal_length']]
new_iris
```

```python
x=new_iris['petal_width']
y=new_iris['petal_length']
```

```python
plt.scatter(x,y)
plt.xlabel('petal_width')
plt.title('scatter plot to show the relationship')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0,)
```

```python
x_train
```

```python
x_train=np.array(x_train).reshape(-1,1)#reshape
```

```python
x_train
```

```python
x_test=np.array(x_test).reshape(-1,1)#reshape the x_test into 2d array
x_test
```

```python
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
model=linear.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
#sns.regplot(x=x,y=y)
```

```python
y_train_pred=linear.predict(x_train)
y_train
```

```python
plt.scatter(x=x_train, y=y_train)
plt.plot(x_train,y_train_pred,color='g')
```
