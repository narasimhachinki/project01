```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
df=pd.read_csv("C:/Users/MADHUSUDHAN/OneDrive/Desktop/Boston.csv")
df
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
correlation_matrix=df.corr().round(2)
#annos=to print the value inside the square
sns.heatmap(data=correlation_matrix,annot=True,annot_kws={'size':8},)
```

```python
features=['rm','lstat']
target=['medv']
```

```python
x=df[features[0]]
y=df[target]
```

```python
plt.scatter(x,y)
plt.xlabel('average number of rooms')
plt.ylabel('median values')
plt.show()
```

```python
x=df[features[1]]
plt.scatter(x,y)
plt.xlabel('lower status of polution')
plt.ylabel('median values')
plt.show()
```

```python
new_x=df[features]
new_y=df[target]
print(new_x.shape)
print(new_y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(new_x,new_y,test_size=0.2,random_state=0)
```

```python
print('x_train',x_train.shape)
print('y_train',y_train.shape)
print('x_test',x_test.shape)
print('y_test',x_test.shape)
```

```python
linear=LinearRegression()
#setting the model or giving training set to model
model=linear.fit(x_train,y_train)
model
```

```python
y_train_pred=model.predict(x_train)
print(y_train_pred)
```

```python
plt.scatter(y_train,y_train_pred)
plt.show()
```

```python
print('mean absolute error',mean_absolute_error(y_train,y_train_pred))
print('mean absolute error',mean_squared_error(y_train,y_train_pred))
print('mean absolute error',np.sqrt(mean_squared_error(y_train,y_train_pred)))
print('r_2 score',r2_score(y_train,y_train_pred))
```
