```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
```

```python
df=pd.DataFrame(data=data.data,columns=data.feature_names)
df
```

```python
df.head()
```

```python
print(data['DESCR'])
```

```python
print(data.keys())
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df)
scaler_data=scaler.transform(df)
scaler_data
```

```python
from sklearn.decomposition import PCA
PCA = PCA()
```

```python
PCA.fit(scaler_data)
```

```python
x_PCA=PCA.transform(scaler_data)
print(scaler_data.shape)
print(x_PCA.shape)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_PCA[:,0],x_PCA[:,1],c=data['target'],cmap='plasma')
plt.xlabel('first principal component',size=20,color='g',family='magneto')
plt.ylabel('second principle component',size=20,color='r',family='magneto')
```
